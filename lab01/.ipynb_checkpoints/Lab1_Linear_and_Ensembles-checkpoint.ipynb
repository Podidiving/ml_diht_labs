### 1. Чтение данных.
Воспользуемся [датасетом](https://archive.ics.uci.edu/ml/datasets/Statlog+%28Vehicle+Silhouettes%29), в котором описываются различные автомобили. Будем решать задачу многоклассовой ($k=4$) классификации.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

dataset = pd.read_csv('car_data.csv', delimiter=',', header=None).values
data = dataset[:, :-1].astype(int)
target = dataset[:, -1]

print(data.shape, target.shape)
```

```python
lr_clf = LogisticRegression(multi_class='ovr', solver='liblinear', max_iter=5000)
```

```python
# And at the best params of the model
grid.best_params_
```

```python
# Save this model
best_lr_clf = grid.best_estimator_
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
 
def show_scores(y_true, y_pred):
    conf_matrix = confusion_matrix(y_true, y_pred)
    accuracy_ = accuracy_score(y_true, y_pred)
    f1_score_ = f1_score(y_true, y_pred, average='macro')
    f1_scores = f1_score(y_true, y_pred, average=None)
    classes = encoder.classes_.shape[0]
    for i in range(classes):
        current_accuracy_score = conf_matrix[i,i] / conf_matrix.sum(axis=1)[i]
        print('Accuracy_score for {:>5s} : {}'.format(encoder.classes_[i],  current_accuracy_score))
    print()
    print('Total accuracy score: {}'.format(accuracy_))
    print()
    
    for i in range(classes):
        print('F1-score for {:>5s} : {}'.format(encoder.classes_[i], f1_scores[i]))
    print()
    print('Average F1-score : {}'.format(f1_score_))
    print()
    
    
def show_roc_curve(y_true, y_score, title=''):
    
    classes = encoder.classes_.shape[0]
    tpr = [0] * classes
    fpr = [0] * classes 
    auc_score = [0] * classes
    
    for i in range(classes):
        current_class = np.array(list(map(lambda x: 1 if x else 0, (y_true == i))))
        y_score_current = y_score[:, i]
        fpr[i], tpr[i], _ = roc_curve(current_class, y_score_current)
        auc_score[i] = roc_auc_score(current_class, y_score_current)
        
    plt.figure(figsize=(16, 9))
    plt.grid(True)
    plt.plot([0, 1], [0, 1], 'ko--')
    plt.title(title, fontsize=15)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    
    for i in range(classes):
        plt.plot(
                    fpr[i],
                    tpr[i],
                    label='ROC curve for {:4s}, roc-auc score = {:0.3f}'
                    .format(encoder.classes_[i], auc_score[i]),
        )
        
    plt.legend(loc='best')    
    plt.show()
```

```python
show_scores(y_test, best_lr_clf.predict(X_test))
show_roc_curve(
                            y_test, 
                            best_lr_clf.decision_function(X_test), 
                            title='Logistic Regression'
)
```

#### 2.6. Bagging.
Перейдем к ансамблям. 

Для построения ансамбля воспользуемся Bagging'ом с числом алгоритмов от 2 до 50. В качестве первого семейства базовых алгоримтов будем использовать линейные модели (т.е. логистическую регрессию), в качестве второго - решающие деревья. 

*Пояснение: Будем строить ансамбль только из моделей из одного семейства, т.е. логистическая регрессия не смешивается с решающими деревьями.*

Для этого можно сгенерировать с помощью метода bootstrap 50 подвыборок из `train` выборки (того же размера, что и исходная), обучить логистическую регрессию и решающее дерево с оптимальными параметрами из предыдущего пункта на каждой из подвыборок и затем усреднить предсказания k моделей.

*Hint: в sklearn может найтись подходящая функция, которая облегчит вам реализацию данного пункта.*

* Постройте график качества классификации и f1-score на `train` и `test` датасетах в зависимости от числа алгоритмов, вошедших в ансамбль.

* Проанализируйте график. Какое количество моделей стоит использовать? Как вы думаете, являются ли параметры решающего дерева, подобранные в предыдущем пункте оптимальными в данном случае?

```python
from sklearn.ensemble import BaggingClassifier
from tqdm import tqdm_notebook

def make_bagging_list(base_estimator, data, target, min_n_estimators=2, max_n_estimators=50):
    bagging_list = []
    for n_estimators in tqdm_notebook(np.arange(min_n_estimators, max_n_estimators + 1)):
        bagging = BaggingClassifier(base_estimator, n_estimators=n_estimators)
        bagging.fit(data, target)
        bagging_list.append(bagging)
    return np.array(bagging_list)
```

```python
def show_f1_and_accuracy_for_bagging(bagging_list, 
                                     data, target, 
                                     data_train, target_train,
                                     min_n_estimators=2, max_n_estimators=50):
    accuracy_list_train = []
    accuracy_list = []
    f1_score_list_average_train = []
    f1_score_list_average = []
    f1_score_list = []
    for bagging in tqdm_notebook(bagging_list):
        pred = bagging.predict(data)
        accuracy_list.append(accuracy_score(target, pred))
        f1_score_list_average.append(f1_score(target, pred, average='macro'))
        f1_score_list.append(f1_score(target, pred, average=None))
        pred_train = bagging.predict(data_train)
        accuracy_list_train.append(accuracy_score(target_train, pred_train))
        f1_score_list_average_train.append(f1_score(target_train, pred_train, average='macro'))
    xs = np.arange(min_n_estimators, max_n_estimators + 1)
    
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    plt.plot(xs, accuracy_list_train, label="train data")
    plt.plot(xs, accuracy_list, label="test data")
    plt.title("Depending the number of estimators on accuracy score")
    plt.xlabel("Estimators")
    plt.ylabel("Accuracy score")
    plt.xticks(xs)
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    plt.plot(xs, f1_score_list_average_train, label="train data")
    plt.plot(xs, f1_score_list_average, label="test data")
    plt.title("Depending the number of estimators on F1 average score")
    plt.xlabel("Estimators")
    plt.ylabel("F1 average score")
    plt.xticks(xs)
    plt.legend()
    plt.show()
    
    f1_score_list = np.array(f1_score_list)
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    for i in range(encoder.classes_.shape[0]):
        plt.plot(xs, f1_score_list[:, i], label="f1 score on {}".format(encoder.classes_[i]))
    plt.title("Depending the number of estimators on F1  score (test data)")
    plt.xlabel("Estimators")
    plt.ylabel("F1 score")
    plt.xticks(xs)
    plt.legend()
    plt.show()
    
```

Начнем с `Logistic Regression`

```python
bagging_list_lr = make_bagging_list(best_lr_clf, X_train, y_train)
```

```python
show_f1_and_accuracy_for_bagging(bagging_list_lr, X_test, y_test, X_train, y_train)
```

Перейдем к беггингу над решащими деревьями

```python
bagging_list_tree = make_bagging_list(best_tree_clf, X_train, y_train)
```

```python
show_f1_and_accuracy_for_bagging(bagging_list_tree, X_test, y_test, X_train, y_train)
```

Беггинг над линейными моделями не приносит почти никакой пользы, что логично, потому что линейная комбинация линейных моделей - линейная модель
Беггинг над линейными деревьями немного улучшает ситуацию, однако видно, что при большом количестве деревьев ошибка на треине становится нулевой - на лицо переобучение.

```python
from sklearn.model_selection import StratifiedKFold
```

```python
n_splits=10
stratified_k_fold = StratifiedKFold(n_splits=n_splits)

X_train_partition = []
y_train_partition = []

for _, test_indices in stratified_k_fold.split(X_train, y_train):
    X_train_partition.append(X_train[test_indices])
    y_train_partition.append(y_train[test_indices])
X_train_partition = np.array(X_train_partition)
y_train_partition = np.array(y_train_partition)
```

```python
def show_scores_on_partial_data(estimator,
                                X_train_partition, y_train_partition, 
                                X_test, y_test,
                                n_splits=10):
    accuracy_score_list_train = []
    accuracy_score_list_test = []
    f1_score_list_average_train = []
    f1_score_list_average_test = []
    f1_score_list = []
    partitions = np.arange(1, n_splits + 1)
    for curr_partition in partitions:
        X_train_curr = np.vstack(X_train_partition[:curr_partition])
        y_train_curr = np.hstack(y_train_partition[:curr_partition])
        
        estimator.fit(X_train_curr, y_train_curr)
        pred_test = estimator.predict(X_test)
        pred_train = estimator.predict(np.vstack(X_train_partition[:]))
        
        accuracy_score_list_test.append(accuracy_score(y_test,
                                                       pred_test)
        )
        accuracy_score_list_train.append(
            accuracy_score(np.hstack(y_train_partition[:]),
                           pred_train)
        )
        f1_score_list_average_test.append(f1_score(y_test, pred_test, 
                                                   average="macro")
        )
        f1_score_list_average_train.append(
            f1_score(np.hstack(y_train_partition[:]), pred_train,
                     average="macro")
        )
        f1_score_list.append(f1_score(y_test, pred_test, average=None))
    
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    plt.plot(partitions, accuracy_score_list_train, label="train data")
    plt.plot(partitions, accuracy_score_list_test, label="test data")
    plt.title("Depending the number of estimators on accuracy score")
    plt.xlabel("Estimators")
    plt.ylabel("Accuracy score")
    plt.xticks(partitions)
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    plt.plot(partitions, f1_score_list_average_train, label="train data")
    plt.plot(partitions, f1_score_list_average_test, label="test data")
    plt.title("Depending the number of estimators on F1 average score")
    plt.xlabel("Estimators")
    plt.ylabel("F1 average score")
    plt.xticks(partitions)
    plt.legend()
    plt.show()
    
    f1_score_list = np.array(f1_score_list)
    plt.figure(figsize=(12, 9))
    plt.grid(True)
    for i in range(encoder.classes_.shape[0]):
        plt.plot(partitions, f1_score_list[:, i], label="f1 score on {}".format(encoder.classes_[i]))
    plt.title("Depending the number of estimators on F1  score (test data)")
    plt.xlabel("Estimators")
    plt.ylabel("F1 score")
    plt.xticks(partitions)
    plt.legend()
    plt.show()
    
```

```python
show_scores_on_partial_data(best_lr_clf, X_train_partition, y_train_partition, X_test, y_test)
```

```python
show_scores_on_partial_data(best_tree_clf, X_train_partition, y_train_partition, X_test, y_test)
```
